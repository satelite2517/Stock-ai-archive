```python
import requests

def get_news(api_key, query, from_date, to_date, language='en', sort_by='publishedAt'):
    url = 'https://newsapi.org/v2/everything'
    params = {
        'q': query,
        'from': from_date,
        'to': to_date,
        'language': language,
        'apiKey': api_key
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        return None

# Replace 'your_api_key' with your actual API key from NewsAPI.org
api_key = '872821aafaac4610b81caa963bede548'
query = 'AOS'
from_date = '2024-07-01'
to_date = '2024-07-26'

news_data = get_news(api_key, query, from_date, to_date)

if news_data:
    articles = news_data.get('articles', [])
    for article in articles:
        print(f"Title: {article['title']}")
        print(f"Description: {article['description']}")
        print(f"URL: {article['url']}\n")
else:
    print("Failed to fetch news articles")
```
